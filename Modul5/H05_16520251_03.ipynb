Deskripsi : visualisasi dasar dataset

```python
import pandas as pd
import matplotlib as plt
```

Menggunakan file dataset nilai_uas.csv sebagai variabel df

```python
df = pd.read_csv("nilai_uas.csv")
```

# 1

menvisualkan histogram nilai kalkulus

```python
df[["nilai_kal"]].plot(kind="hist",bins=[0,20,40,60,80,100],rwidth=0.9)
```

# 2

Visualisasi diagram batang jumlah mahasiswa tiap fakultas secara horizontal

```python
# mendeclare count sebagai data baru yang berisi jumlah mahasiswa tiap fakultas
count_fakultas = df["fakultas"].value_counts()
# menvisualkan data count dengan diagram batang horizontal
count_fakultas.plot(kind="barh",title="Jumlah Mahasiswa Tiap Fakultas")
```

# 3

Visualisasi diagram pie jumlah mahasiswa tiap fakultas

```python
# menvisualkan data count dengan diagram pie
count_fakultas.plot(kind="pie")
```

# 4

Melihat dari kedua diagram di atas, mahasiswa terbanyak dapat terlihat berada di fakultas STEI dengan jumlah mahasiswa hampir 500 individu

Diagram batang lebih baik dalam menampilkan jumlah mahasiswa tiap fakultas, karena disitu terlihat dengan jelas perbedaan antar fakultas dengan adanya angka. Sedangkan di diagram pie tidak terdapat angka yang menyebabkan kecenderungan yang sulit dalam membedakan besar kecilnya jumlah mahasiswa antar fakultas jika hanya dilihat dari besar kecilnya bentuk potongan lingkaran saja.
